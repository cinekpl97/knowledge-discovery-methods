## Redukcja wymiarowoci

Praca z wielowymiarowymi danymi niesie ze sob rne ograniczenia. W jaki sposb moemy wywietli prbki, ktre maj $4$, $5$, $6$, $7$ i wicej cech? Nie moemy stworzy przecie wykresu $4D$, czy $7D$. Wizualizacja takich danych jest problemem. Kiedy mamy wiele cech w zestawie danych, wpywa to rwnie na szybko dziaania naszego modelu. Takie dane zajmuj wicej miejsca na nonikach pamici. Jeli mamy zbir, ktry zawiera $1000$ cech, to moemy si zastanawia, czy na pewno wszystkie cechy s istotne i potrzebne dla naszego modelu. Moe wystarczy jedynie $100$ cech i wyniki dziaania modelu uczenia maszynowego byyby rwnie satysfakcjonujce, a przy okazji zmniejszylibymy wymiarowo prbek i czas oblicze. 

Problem redukcji wymiarowoci jest powszechny w tematyce uczenia maszynowego. Istniej dwa gwne podziay takiej redukcji - selekcja i ekstrakcja cech. Selekcja cech polega na wybraniu pewnego podzbioru cech z oryginalnego zestawu wszystkich cech. W takim przypadku zakadamy, e nie wszystkie cechy s nam potrzebne do stworzenia dobrego modelu, wic pozbywamy si czci z nich, liczc na to, e wyniki nie ulegn pogorszeniu. Selekcja cech pozostawia cechy w swojej oryginalnej postaci, nie modyfikuje ich wartoci, zmieniamy tylko ich ilo. Ekstrakcja cech jest scenariuszem nieco bardziej skomplikowanym, poniewa polega na redukcji cech poprzez przeksztacenie oryginalnego zestawu cech do nowej przestrzeni cech, w ktrej pozostae cechy nie maj swoich pierwotnych wartoci. Obie kategorie maj swoje wady i zalety. Dzisiaj skupimy si na algorytmie z rodziny ekstrakcji cech. 

##  Analiza gwnych skadowych

Jedn z najpopularniejszych metod ekstrakcji cech jest **Analiza gwnych skadowych (ang. Principal component analysis)** w skrcie **PCA**. Jest to algorytm, ktry redukuje wymiarowo danych poprzez rzutowanie ich na przestrze o mniejszej liczbie wymiarw. Rzutowanie to nastpuje poprzez znalezienie kierunkw, w ktrych jest najwiksza wariancja danych. Zaoeniem algorytmu jest zatem znalezienie nowej przestrzeni cech, tak aby zachowa jak najwicej informacji zawartych w oryginalnych cechach. Przed omwieniem szczegw algorytmu zobaczmy jego zasad dziaania.

Majc dwuwymiarowy zestaw danych z cechami $x_1$ i $x_2$, chcielibymy zredukowa go do postaci jednowymiarowej. 



Aby to zrobi, musimy znale kierunki, w ktrych dane s najbardziej zrnicowane (maj najwiksz wariancj). Takie kierunki bd dwa, a dodatkowo bd one zawsze do siebie prostopade. Te kierunki, to wanie gwne skadowe.



Majc wyznaczone gwne skadowe, moemy zauway, e wzdu prostej z czerwon strzak jest znacznie wiksza wariancja danych. Jeli zatem chcielibymy zredukowa przestrze z $2D$ na $1D$, to powinnimy rzutowa wszystkie punkty z dwuwymiarowej przestrzeni na prost z czerwon strzak.

Po zrzutowaniu punktw na t prost dane prezentuj si nastpujco:



Moemy te sprbowa zrzutowa punkty na prost z niebiesk strzak, eby przekona si, e punkty bd bardziej "zbite" (mniejsza wariancja danych).



Ta idea dla prostego przypadku dwuwymiarowego jest rozszerzalna i dziaa dokadnie tak samo dla danych wielowymiarowych. Jeli mamy ju intuicj, w jaki sposb redukowane s wymiary w algorytmie PCA, przejdmy do szczegw dziaania algorytmu.

## Algorytm PCA

Na pocztku stwrzmy zbir danych, na ktrym bdziemy przeprowadza obliczenia.

| **x1** | **x2** |
|--------|--------|
| 1\.2   | 1\.5   |

````python
X = np.array([
    [1.2, 1.5],
    [0.2, 0.1],
    [1.6, 1.8],
    [2.2, 2.5],
    [0.5, 0.7],
    [3.0, 2.6],
    [3.2, 3.4],
    [0.8, 1.0],
    [2.6, 2.8],
    [1.5, 1.4]])
````

### Wycentrowanie danych

Pierwszym etapem jest wycentrowanie zestawu danych, to znaczy odjcie redniej od kadej prbki. Obliczamy redni dla cechy $x_1$ i cechy $x_2$.

````python
X_centered = X - X.mean(axis=0)

X_centered
````

````python
plt.scatter(X_centered[:, 0], X_centered[:, 1], marker='+')
````

### Obliczenie macierzy kowariancji

Po wycentrowaniu zestawu danych naley obliczy macierz kowariancji zgodnie ze wzorem:


<!-- $$\Sigma = \frac{1}{m-1}\sum_{i=1}^{n}(x^{(i)})(x^{(i)})^T$$ -- DO POPRAWY!!! -->
$$\Sigma = \frac{1}{m-1}X^TX$$

gdzie:

* $X^T$ jest macierz o wymiarach $nxm$

* $X$ jest macierz o wymiarach $mxn$
<!-- * $x^{(i)}$ jest wektorem o wymiarach $nx1$ -->

<!-- * $(x^{(i)})^T$ jest wektorem o wymiarach $1xn$ -->

Wic w wyniku mnoenia powstaje macierz $\Sigma$ o wymiarach $nxn$.


````python
m = X.shape[0]
X_cov = np.dot(X_centered.T, X_centered) / (m - 1)

X_cov
````

### Obliczenie wektorw i wartoci wasnych

Ten etap jest sercem metody analizy gwnych skadowych. Wektory wasne to kierunki, w ktrych jest najwiksza wariancja danych. Wartoci wasne, ktre skojarzone s z wektorami wasnymi, okrelaj ich istotno. Oznacza to, e wektor wasny z najwiksz wartoci wasn jest pierwsz gwn skadow, wektor wasny z drug najwiksz wartoci jest drug gwn skadow i tak dalej. 

Wektory i wartoci wasne mona obliczy za pomoc metody `np.linalg.eig()` podajc jako parametr obliczon wczeniej macierz kowariancji.

````python
eigenvalues, eigenvectors = np.linalg.eig(X_cov)
eigenvalues = eigenvalues.real
eigenvectors = eigenvectors.real
````

Bardzo istotne jest to, e wektory wasne zwracane przez `np.linalg.eig()` s reprezentowane jako **kolumny**, a nie wiersze. Oznacza to, e pierwsz kolumn jest pierwszy wektor wasny, drug kolumn jest drugi wektor wasny itp.

Wane jest rwnie to, e zwrcone wartoci wasne niekoniecznie musz by posortowane. Musimy zatem sami posortowa wartoci wasne (zachowujc powizanie z wektorami wasnymi), eby mie pewno, e s ustawione malejco.

````python
idx = eigenvalues.argsort()[::-1]
eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:,idx]

eigenvalues, eigenvectors
````

### Rysowanie danych z gwnymi skadowymi

Majc wektory wasne moemy w przypadku dwuwymiarowym do atwo je narysowa. Obliczamy nachylenie zgodnie ze wzorem $\frac{\Delta y}{\Delta x}$.

````python
plt.scatter(X_centered[:, 0], X_centered[:, 1], marker='+')

line = np.linspace(-1, 1, 100)
plt.arrow(x=0, y=0, dx=eigenvectors[0, 0], dy=eigenvectors[1, 0], head_width=0.1, head_length=0.1, fc='r', ec='k')
plt.arrow(x=0, y=0, dx=eigenvectors[0, 1], dy=eigenvectors[1, 1], head_width=0.1, head_length=0.1, fc='b', ec='k')
plt.plot(line, line * (eigenvectors[1, 0]/eigenvectors[0, 0]), color='black')
plt.plot(line, line * (eigenvectors[1, 1]/eigenvectors[0, 1]), color='black')

plt.xlabel("x1")
plt.ylabel("x2")
plt.grid()
plt.show()
````

### Rzutowanie danych na k gwnych skadowych

Majc obliczon macierz z kolejnymi wektorami wasnymi, moemy wykorzysta j do zredukowania oryginalnej iloci wymiarw. W sytuacji, gdy mamy jedynie dwa wymiary w oryginalnym zestawie danych, nie mamy duego pola do popisu. Jednak w zdecydowanej wikszoci przypadkw bdziemy pracowa z danymi, ktre maj znacznie wicej cech. Dziki temu moemy wybra $k$ pierwszych gwnych skadowych i za ich pomoc rzutowa dane do przestrzeni o $k$-tej liczbie wymiarw. 

Operacja rzutowania opisana jest za pomoc wzoru:

$$z = U_{reduced}^T * x$$

gdzie:
* $U_{reduced}$ oznacza macierz z k pierwszymi wektorami wasnymi - wymiar $nxk$
* $x$ to oryginalna prbka - wymiar $nx1$

Zatem wyjciowa prbka $z$ bdzie miaa wymiar $kx1$.

Moemy rwnie rzutowa cay zbir danych "na raz" i wtedy wzr bdzie mia nastpujc form:

$$Z = X * U_{reduced}$$

gdzie:
* $X$ to wycentrowany zbir danych - wymiar $mxn$
* $U_{reduced}$ to macierz z k pierwszymi wektorami wasnymi - wymiar $nxk$

Zatem nowy zbir danych ze zredukowan iloci wymiarw bdzie mia wymiar $mxk$.

#### Rzutowanie danych na pierwsz gwn skadow

````python
X_pca_1 = np.dot(X_centered, eigenvectors[:, 0])
plt.scatter(X_pca_1, np.zeros(10), marker='+')
plt.xlim((-1.5, 2))
plt.xlabel("z")
plt.show()
````

#### Rzutowanie danych na drug gwn skadow

````python
X_pca_2 = np.dot(X_centered, eigenvectors[:, 1])
plt.scatter(X_pca_2, np.zeros(10), marker='+')
plt.xlim((-1.5, 2))
plt.xlabel("z")
plt.show()
````

Wida, e znacznie korzystniejsze jest rzutowanie danych na pierwsz gwn skadow, poniewa dane nie s tak zbite, jak w przypadku drugiej gwnej skadowej.

### Powrt do oryginalnej liczby wymiarw

Po zrzutowaniu danych na przestrze o mniejszej iloci wymiarw jestemy w stanie powrci do oryginalnego $n$-wymiarowego zestawu danych, wykorzystujc wszystkie gwne skadowe. Operacja ta jednak powoduje strat pewnej iloci informacji, ktra zostaa utracona poprzez operacje rzutowania. 

Aby powrci do oryginalnej liczby wymiarw, naley wykona nastpujc operacj:

$$x_{reconstructed} = U_{reduced} * z$$


* $U_{reduced}$ ma wymiary $nxk$
* $z$ ma wymiary $kx1$ 
* zatem $x_{reconstructed}$ bdzie miao wymiar $nx1$.

Posta macierzowa:

$$X_{reconstructed} = Z * U_{reduced}^T$$

* $Z$ ma wymiary $mxk$
* $U_{reduced}^T$ ma wymiary $kxn$
* zatem $X_{reconstructed}$ bdzie miao wymiar $mxn$.

#### Operacja odwrotna po rzutowaniu danych na pierwsz gwn skadow

````python
old_data_from_pca_1 = np.dot(X_pca_1.reshape(X_pca_1.shape[0], 1), eigenvectors[:, :1].T)
plt.scatter(old_data_from_pca_1[:, 0], old_data_from_pca_1[:, 1], marker='+')
plt.xlim((-1.5, 2))
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
````

Wida, e z atwoci udao si odtworzy dane idce wzdu pierwszej gwnej skadowej, ale nie wzdu drugiej.

### Wspczynnik wariancji wyjanionej

Gdy analizujemy prosty przypadek z dwoma wymiarami, nie mamy wielkiego wyboru jeli chodzi o redukcje cech. Moemy jedynie rzutowa prbki na przestrze jednowymiarow. W sytuacji, gdy mamy wiele wymiarw problemem moe okaza si wybr waciwej iloci gwnych skadowych, ktre bd suyy do stworzenia nowej przestrzeni prbek. Przykadowo, jeli mamy 6000 wymiarw, to ile $k$ pierwszych gwnych skadowych powinnimy wzi? 2000? 1000? a moe jedynie 100? Moglibymy oczywicie prbowa wszystkich wartoci z pewnego przedziau, a nastpnie mierzy rnice pomidzy oryginalnymi prbkami i prbkami zrekonstruowanymi. To daoby nam pewn miar bdu. Odpowiednie przeksztacenia pozwoliyby nawet zawrze t wartoci w przedziale $(0,1)$. Jednak takie rozwizanie wymagaoby bardzo duej iloci oblicze. 

Na szczcie istnieje inny sposb. Obliczone wraz z wektorami wasnymi, wartoci wasne maj wan waciwo. Informuj o tym, jaka ilo wariancji niesiona jest za pomoc skojarzonej gwnej skadowej. Jest to tak zwana **wariancja wyjaniona**. Aby wycign t informacj, naley obliczy iloraz kadej wartoci wasnej i sumy wszystkich wartoci wasnych:

$$\frac{eigenvalue^{(i)}}{\sum_{i=1}^{n}eigenvalue^{(i)}}$$

Przykadowo, jeeli wariancja wyjaniona dla pierwszej waroci wasnej ma warto $0.9$, oznacza to, e a $90%$ wariancji jest wyjanionej za pomoc pierwszej gwnej skadowej. Czyli pozostae $10%$ jest zawarte w pozostaych gwnych skadowych.

Czciej jednak stosowana jest operacja obliczania wartoci wyjanionej z wykorzystaniem skumulowanej sumy. Taka operacja pozwala na atwe sprawdzenie, jaka ilo wariancji jest wyjaniona za pomoc $k$ pierwszych skadowych.

$$\frac{\sum_{i=1}^{k} eigenvalue^{(i)}}{\sum_{i=1}^{n}eigenvalue^{(i)}}$$

Dziki temu moemy ustali, e chcemy, aby nowa przestrze cech zachowaa $95%$ wariancji:

$$\frac{\sum_{i=1}^{k} eigenvalue^{(i)}}{\sum_{i=1}^{n}eigenvalue^{(i)}} \geq 0.95$$

````python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt


def covariance(X_centered, img):
    return np.dot(X_centered.T, X_centered) / (img.shape[0] - 1)

def get_eigenvalues_and_eigenvectors(X_cov):
    eigenvalues, eigenvectors = np.linalg.eig(X_cov)

    idx = eigenvalues.argsort()[::-1]
    eigenvalues = eigenvalues[idx]
    eigenvectors = eigenvectors[:, idx]
    
    eigenvalues = eigenvalues.real
    eigenvectors = eigenvectors.real

    return eigenvalues, eigenvectors

def center(img):
    X_centered = []
    transposed_img = img.T
    for column in transposed_img:
        mean = (np.sum(column)) / img.shape[0]
        X_centered.append(column - mean)

    X_centered = np.array(X_centered)
    X_centered = X_centered.T
    return X_centered

def construct(k_array, X_pca_table, eigenvectors):
    X_recon_table = [0] * len(k_array)
    for index, X_rec in enumerate(X_recon_table):
        X_recon_table[index] = np.float_(np.dot(X_pca_table[index], (eigenvectors[:, : k_array[index]]).T))
        X_recon_table[index] = (X_recon_table[index] + np.absolute(X_recon_table[index].min()))
        X_recon_table[index] = (X_recon_table[index] / X_recon_table[index].max()) * 255

    return X_recon_table


def sum_of_eigenvalues(eigenvalues, img_array):
    sum_all_eigenvalues = np.sum(np.float_(eigenvalues))
    sums_of_k_eigenvalues = np.zeros(img_array.shape[1])

    sums_of_k_eigenvalues[0] = np.float_(eigenvalues[0])
    for i in range(1, img_array.shape[1]):
        sums_of_k_eigenvalues[i] = np.sum(np.float_(eigenvalues[: i + 1]))

    sums = []
    for i in range(len(sums_of_k_eigenvalues)):
        sums.append(sums_of_k_eigenvalues[i] / sum_all_eigenvalues)

    return sum_all_eigenvalues, sums


def pca():
    file_name = 'dolphines.jpg'
    img = PIL.Image.open(file_name).convert('L')
    img = np.array(img)
    X_centered = center(img)
    X_cov = covariance(X_centered, img)
    eigenvalues, eigenvectors = get_eigenvalues_and_eigenvectors(X_cov)
    k_array = [round(img.shape[1] / 4), round(img.shape[1] / 8), round(img.shape[1] / 16),
                  round(img.shape[1] / 32), round(img.shape[1] / 64), round(img.shape[1] / 128), 2]

    X_pca_array = [0] * len(k_array)
    for index, X_pca in enumerate(X_pca_array):
        X_pca_array[index] = np.float_(np.dot(X_centered, eigenvectors[:, : k_array[index]]))
    X = construct(k_array, X_pca_array, eigenvectors)
    sum_all_eigenvalues, sums = sum_of_eigenvalues(eigenvalues, img)

    x = np.array(range(len(sums)))
    y = sums
    x_value = np.interp(0.95, y, x)
    plt.plot(x, y, c='k', linestyle='solid')
    plt.plot(x, np.ones(len(sums)) * 0.95)
    plt.show()

    print('95% wariancji dla ' + str(np.int(np.ceil(x_value))) + ' pierwszych skadowych')

pca()


````

## Zadanie 2

Obrazy kolorowe zapisane s w formacie RGB. Oznacza to, e taki obraz ma 3 kanay - R, G i B. Przerb program z zadania pierwszego w taki sposb, aby dokonywa kompresji zdjcia w formacie RGB (usunicie funkcji `.convert('L')`). Aby dokona takiej kompresji, naley przeprowadzi kompresje algorytmem PCA osobno dla kadego kanau, a nastpnie poczy kanay razem, tak by tworzyy oryginalny obraz. Wykonaj eksperymenty dla rnych wartoci parametru $k$.

````python
from PIL import Image, ImageDraw, ImageFont
import numpy as np

def get_eigenvalues_and_eigenvectors(X_cov):
    eigenvalues, eigenvectors = np.linalg.eig(X_cov)
    idx = eigenvalues.argsort()[::-1]
    eigenvalues = eigenvalues[idx]
    eigenvectors = eigenvectors[:, idx]
    eigenvalues = eigenvalues.real
    eigenvectors = eigenvectors.real
    return eigenvalues, eigenvectors

def covariance(X_centered, img):
    return np.dot(X_centered.T, X_centered) / (img.shape[0] - 1)

def pca(k, X_centered, eigenvectors):
    return np.float_(np.dot(X_centered, eigenvectors[:, : k]))

def center(img):
    X_centered = []
    for column in img.T:
        mean = (np.sum(column)) / img.shape[0]
        X_centered.append(column - mean)

    X_centered = np.array(X_centered)
    X_centered = X_centered.T
    return X_centered


def sums(eigenvalues, img):
    eigenvalues_total_sum = np.sum(np.float_(eigenvalues))
    k_sum = np.zeros(img.shape[1])
    k_sum[0] = np.float_(eigenvalues[0])
    for i in range(1, img.shape[1]):
        k_sum[i] = np.sum(np.float_(eigenvalues[: i + 1]))
    sums = []
    for i in range(len(k_sum)):
        sums.append(k_sum[i] / eigenvalues_total_sum)

    return eigenvalues_total_sum, sums


def global_pca(img, k):
    X_centered = center(img)
    X_cov = covariance(X_centered, img)
    eigenvalues, eigenvectors = get_eigenvalues_and_eigenvectors(X_cov)
    X_pca = pca(k, X_centered, eigenvectors)
    X = np.float_(np.dot(X_pca, (eigenvectors[:, : k]).T))
    X = (X + np.absolute(X.min()))
    X = (X / X.max()) * 255

    return X


def pca_function(k):
    file_name = 'dolphines.jpg'
    img = Image.open(file_name)
    img = np.array(img)

    RGB_array = []
    for sheet in np.transpose(img, [2, 0, 1]):
        RGB_array.append(global_pca(sheet, k))
    RGB_array = np.array(RGB_array)
    RGB_array = np.transpose(RGB_array, [1, 2, 0])
    RGB_array = np.int_(RGB_array)
    img_to_print = Image.fromarray(RGB_array.astype(np.uint8))
    draw = ImageDraw.Draw(img_to_print)
    img_to_print.show()

chosen_values = [640, 320, 160, 80, 40, 20, 10]
for i in chosen_values:
    pca_function(i)
````

````python
from sklearn import datasets
from sklearn.preprocessing import StandardScaler

def center(data):
    X_centered = []
    for column in data.T:
        mean = (np.sum(column)) / data.shape[0]
        X_centered.append(column - mean)

    X_centered = np.array(X_centered)
    X_centered = X_centered.T
    return X_centered


def covariance(X_centered, data):
    return np.dot(X_centered.T, X_centered) / (data.shape[0] - 1)


def get_eigenvalues_and_eigenvectors(X_cov):
    eigenvalues, eigenvectors = np.linalg.eig(X_cov)
    idx = eigenvalues.argsort()[::-1]
    eigenvalues = eigenvalues[idx]
    eigenvectors = eigenvectors[:, idx]
    eigenvalues = eigenvalues.real
    eigenvectors = eigenvectors.real

    return eigenvalues, eigenvectors

def sums(eigenvalues, data):
    sum_total = np.sum(np.float_(eigenvalues))
    k_sums = np.zeros(input.shape[1])

    k_sums[0] = np.float_(eigenvalues[0])
    for i in range(1, data.shape[1]):
        k_sums[i] = np.sum(np.float_(eigenvalues[: i + 1]))

    sums = []
    for i in range(len(k_sums)):
        sums.append(k_sums[i] / sum_total)

    return sum_total, sums


def pca_single(data, k):
    X_centered = center(data)
    x_cov = covariance(X_centered, data)
    eigenvalues, eigenvectors = get_eigenvalues_and_eigenvectors(x_cov)
    single_pca_val = np.float_(np.dot(X_centered, eigenvectors[:, : k]))
    x = np.float_(np.dot(single_pca_val, (eigenvectors[:, : k]).T))

    return x


def pca(data, k):
    scaler_x = StandardScaler()
    data = scaler_x.fit_transform(data)
    x = pca_single(data, k)

    return x

colors = {0: 'red', 1: 'blue', 2: 'green'}
iris = datasets.load_iris()
X = pca(iris.data, 2)
x_1 = np.array(X[:, 0])
x_2 = np.array(X[:, 1])
cats = np.array(iris.target)
fig, ax = plt.subplots()
for j in np.unique(cats):
    i = np.where(cats == j)
    ax.scatter(x1[i], x2[i], c=colors[j])
ax.legend()
plt.show()

scaler = StandardScaler()
data = scaler.fit_transform(iris.data)
x_1 = np.array(data[:, 0])
x_2 = np.array(data[:, 1])
cats = np.array(iris.target)
fig, ax = plt.subplots()
for j in np.unique(cats):
    i = np.where(cats == j)
    ax.scatter(x_1[i], x_2[i], c=colors[j])
ax.legend()
plt.show()

x_3 = np.array(data[:, 2])
x_4 = np.array(data[:, 3])
cats = np.array(iris.target)
fig, ax = plt.subplots()
for j in np.unique(cats):
    i = np.where(cats == j)
    ax.scatter(x_3[i], x_4[i], c=colors[j])
ax.legend()
plt.show()
````
